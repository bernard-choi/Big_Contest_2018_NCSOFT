```python
##길드지수 뽑아내기
import pandas as pd
train_guild = pd.read_csv("train_guild.csv")
test_guild = pd.read_csv("test_guild.csv")
total_guild = pd.concat([train_guild,test_guild])
yunsik  = []
yunsik.extend([x.split(',')  for  x in  total_guild['guild_member_acc_id']]) ## ,로 나눠준다. 
total_guild['new_variable'] = yunsik
from itertools import chain
yunsik2 = list(chain.from_iterable(yunsik))
aa=[]
for i in range(len(total_guild)):
    for j in range(len(total_guild.iloc[i,2])):
        aa.append(total_guild.iloc[i,0])
yeah = pd.DataFrame({'guild_id':aa,'guild_member':yunsik2})
yeah = yeah.sort_values(['guild_member'])
yeah = yeah.reset_index()
yeah = yeah[['guild_id','guild_member']]
yeah.drop_duplicates(inplace=True)


totalset = pd.read_csv('korea.csv')
totalset_for_guild = totalset[['acc_id','cnt_enter_raid',]] ## 여기가 cnt_enter_raid면 레이드 인덱스, 
guildguild = totalset_for_guild.groupby(['acc_id']).sum().reset_index()


i=0
cc=[]
while i < len(yeah)-12:
        aa=[]
        for j in range(i,i+12):
            if yeah.iloc[i,1] == yeah.iloc[j,1]:
                aa.append(yeah.iloc[j,0])

        i+=len(aa)
        cc.append(aa)

while i < len(yeah):
    aa=[]
    for j in range(i,len(yeah)):
        if yeah.iloc[i,1] == yeah.iloc[j,1]:
            aa.append(yeah.iloc[j,0])
    i += len(aa)
    cc.append(aa)

nunu=[len(x) for x in cc]

guild_member= list(yeah['guild_member'].unique())
yunsik = pd.DataFrame({'acc_id':guild_member,"guild_id":cc,"guild_count":nunu})


guild_member_number = []
for i in range(len(total_guild)):
    guild_member_number.append(len(total_guild.iloc[i,2]))
total_guild['guild_member_number'] = guild_member_number


yeah.columns = ['guild_id','acc_id']
for_merge = pd.merge(yeah,guildguild,on='acc_id',how='left')
for_merge.fillna(0,inplace=True)
for_merge2 = for_merge.groupby(['guild_id']).mean().reset_index()
total_dictionary4 = for_merge2.set_index('guild_id').T.to_dict()
final = pd.read_csv('korea.csv')
final.drop(['Unnamed: 0'],axis=1,inplace=True)
final2 = final[final['guild_count']!=0]
final2 = pd.DataFrame(final2).reset_index()
aa=[]
cc=[]
for i in range(len(final2)):
    bb=final2['guild_id'][i].replace("'","").replace("]","").replace("\"","").replace('[','').replace(",","").split(' ')
    cc.append(bb)
ff=[]
for i in range(len(cc)):
        ff.append([total_dictionary4.get(n,n) for n in cc[i]])
aa=[]
for i in range(len(ff)):
    cc=[]
    for j in range((len(ff[i]))):
        cc.extend(ff[i][j].values())
    aa.append(cc)
bb = [max(x) for x in aa]



yunsikus = pd.DataFrame({'acc_id':final2['acc_id'],'guild_raid_index':bb})
aaaa = yunsikus.drop_duplicates()
cniposc = pd.merge(final[['acc_id']],aaaa,on="acc_id",how='left')
cniposc.fillna(0,inplace=True)
cniposc.drop_duplicates(inplace=True)

```
